```python
#As we can see there are alot of outliers in these columns. To delete these outliers we will use IQR 

df3 = df.loc[:,['price','latitude','longitude','baths','bedrooms','area_foot']] 

def find_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers

for i in df3:
    outliers = find_outliers_IQR(df3[i])
    print("name of column: "+ i)
    print("number of outliers: "+ str(len(outliers)))
    print("max outlier value: "+ str(outliers.max()))
    print("min outlier value: "+ str(outliers.min()) + "\n")
```

```python
df3 = df.loc[:,['price','latitude','longitude','baths','bedrooms','area_foot']]

def drop_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    outliers_dropped = outliers.dropna().reset_index(drop=True)
    return outliers_dropped

df3 = drop_outliers_IQR(df3)
df3
```

# Formulating at least 3 hypotheses about this data

Hypothesis 1

```python
import statsmodels.formula.api as smf
model = smf.ols(formula='price ~ location', data=df).fit()

print(model.summary())
```

Firstly, check the R-squared value which measures how much of the variation in the dependent variable (price) is explained by the independent variables (location). In this case, the R-squared value is 0.016 which indicates that the location variable explains only 1.6% of the variation in the price.

Next, check the coefficient values (coef) of the location variable. The intercept value represents the price when the location is not considered. The coefficient values for each location indicate how much the price changes when the location changes, holding other variables constant.

The P-value (P>|t|) column represents the probability of obtaining the observed result (or more extreme) if the null hypothesis is true. In this case, the null hypothesis is that the location variable has no effect on the price. A P-value less than 0.05 indicates that the null hypothesis can be rejected at the 5% level of significance.

In the output provided, most of the P-values for the location variables are greater than 0.05, indicating that the null hypothesis cannot be rejected. Therefore, the location variable does not have a significant effect on the house prices based on this model.

```python
model = smf.ols(formula='price ~ property_type', data=df).fit()

print(model.summary())
```

This is the output of an Ordinary Least Squares (OLS) regression model, where the dependent variable is "price" and the independent variable is "property_type". The model tries to estimate the relationship between the property type and the house price. The table provides information on the estimated coefficients, their standard errors, t-values, and p-values.

The null hypothesis for each coefficient is that the coefficient is equal to zero, indicating that the property type does not have a significant effect on the house price. The alternative hypothesis is that the coefficient is not equal to zero, indicating that the property type has a significant effect on the house price.

In this case, the p-values for all property types are greater than 0.05, which means we fail to reject the null hypothesis and conclude that property type does not have a statistically significant effect on house price. The R-squared value of 0 indicates that the model explains none of the variation in house price, given the property type.

```python
import pandas as pd
import scipy.stats as stats

lhr_n1_wapdatown = df.loc[df.location=='Wapda Town',:].reset_index(drop=True)
lhr_n2_johartown = df.loc[df.location=='Johar Town',:].reset_index(drop=True)

group1 = lhr_n1_wapdatown.price
group2 = lhr_n2_johartown.price

t, p = stats.ttest_ind(group1, group2, equal_var=True)

print('t =', t)
print('p-value =', p)

alpha = 0.05
if p < alpha:
    print('Reject null hypothesis')
else:
    print('Fail to reject null hypothesis')
```
